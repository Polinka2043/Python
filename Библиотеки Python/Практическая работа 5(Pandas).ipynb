```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Create a dataframe
#np.random.seed(5)
#df = pd.DataFrame(np.random.randn(100, 4), columns=list('ABCD'))
film=pd.read_csv("film.csv",delimiter=',')
film
```

```python
import pylab
pylab.subplot (2, 2, 1)
x=np.linspace(1980,2030,10)
pylab.plot(x,film["Number of ratings"])
pylab.title ("Линейный график")

pylab.subplot (2, 2, 2)
types=['Fantasy','Action','Comedy','Crime','Drama']
df1 = film.set_index('Year')['Genre'].str.get_dummies(',')[types].sum()
df1.plot.bar()    
pylab.title ("Гистограмма значений")


pylab.subplot (2, 2, 3)
pylab.barh(film['Film'],film['Estimation'])
plt.title("Гистограмма значений оценок",x= 0.5 , y=-0.3)
```
