```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Using a inbuilt style to change 
# the look and feel of the plot
plt.style.use("fivethirtyeight")
 
# setting figure size to 12, 10
plt.figure(figsize=(20, 20))
 
```

```python
film=pd.read_csv("film1.csv",delimiter=',')
film
```

<h6>Линейный график измения количества отзывов<h6>

```python
fig, ax = plt.subplots()
x=np.linspace(1980,2030,10)
ax.plot(x,film["Number of ratings"])
```

<h6>Столбчатая диаграмма по жанрам<h6>

```python
types=['Fantasy','Action','Comedy','Crime','Drama']
df1 = film.set_index('Year')['Genre'].str.get_dummies(',')[types].sum()
df1.plot.bar()
```

<h6>Сортирую по оценке<h6>

```python
film=film.sort_values(by=['Estimation'],ascending=False)
```

```python
plt.figure(figsize=(20, 10))
fig, ax = plt.subplots()
ax.barh(film['Film'],film['Estimation'])
```
